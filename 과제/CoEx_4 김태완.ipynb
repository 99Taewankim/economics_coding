The data for this figure are provided below. The variables `male` and `female` contain the male and female population sizes in 21 age groups of 5 year interval. The data are taken from the KOSIS.

```python
# population data of Korea in 2021

age_group = ['0~4 years', '5~9 years', '10~14 years', '15~19 years', '20~24 years', '25~29 years', '30~34 years', '35~39 years', '40~44 years', '45~49 years', '50~54 years', '55~59 years', '60~64 years', '65~69 years', '70~74 years', '75~79 years', '80~84 years', '85~89 years', '90~94 years', '95~99 years', '100 and over']
male = [805852, 1133426, 1186319, 1198911, 1681183, 1954718, 1759768, 1877724, 2038116, 2100431, 2243979, 2064943, 2016321, 1433783, 975415, 685723, 447622, 189457, 47148, 8149, 1056]
female = [764557, 1078946, 1118371, 1114388, 1542471, 1708409, 1572504, 1744623, 1943594, 2033792, 2233654, 2032973, 2081537, 1542824, 1114229, 898881, 731214, 428224, 161610, 35719, 5507]
```

```python
# your code here
import matplotlib.pyplot as plt
import numpy as np

# x  
x = np.arange(len(age_group))
male=[round(i,-3) for i in male]
female=[round(i,-3) for i in female]

#   
fig, ax = plt.subplots(figsize=(10, 6))  #   
ax.barh(x, male, color='blue', label='Male')
ax.barh(x, [-val for val in female], color='red', label='Female')

#     (male)
for i in range(len(age_group)):
    ax.text(male[i], i, f'{male[i]}', va='center', color='black', fontsize=8, ha='left')

#     (female)
for i in range(len(age_group)):
    ax.text(-female[i], i, f'{female[i]}', va='center', color='black', fontsize=8, ha='right')

#   
ax.set_title('Population Pyramid of Korea in 2021')

#  
ax.get_xaxis().set_visible(False)

# y  
ax.set_yticks(x)
ax.set_yticklabels(age_group)

#  
ax.legend()

#  
plt.tight_layout()
plt.show()
```

- income_total: the average amount of total income for each income decile from 1 to 10. The unit is 10,000 KRW.
- income_disp: the average amount of disposable income for each income decile from 1 to 10. The unit is 10,000 KRW.
- income_no_cons: the average amount of non-consumption income (taxex, social insurance controbutions, etc.) for each income decile from 1 to 10. The unit is 10,000 KRW.
- age_head: the average age of the head of household. 
- asset_total: the average amount of total assets for each income decile from 1 to 10. The unit is 10,000 KRW.
- asset_financial: the average amount of financial assets for each income decile from 1 to 10. The unit is 10,000 KRW.
- asset_real: the average amount of real assets for each income decile from 1 to 10. The unit is 10,000 KRW.
- debt: the average amount of total debt for each income decile from 1 to 10. The unit is 10,000 KRW.


```python
# data from household survey

income_total = [897, 1749, 2624, 3537, 4491, 5582, 6863, 8436, 10909, 19042]
income_disp = [797, 1544, 2255, 3008, 3810, 4679, 5753, 6992, 8884, 14563]
income_no_cons = [101, 205, 370, 529, 681, 903, 1109, 1444, 2025, 4478]
age_head = [68.2, 63.1, 57.3, 54.6, 53.2, 51.5, 51.4, 50.1, 50.9, 51.5]
asset_total = [12407, 21970, 27511, 32508, 41889, 45692, 57223, 66604, 86386, 155475]
asset_real = [9484, 17562, 21487, 24528, 33018, 34527, 44277, 52475, 67519, 121542]
asset_financial = [2922, 4409, 6024, 7980, 8871, 11165, 12947, 14129, 18867, 33933]
debt = [1068, 2364, 3975, 5042, 6969, 8346, 9954, 12688, 15637, 25644 ]
```

```python
# your code here
import matplotlib.pyplot as plt
import numpy as np

# data from household survey

income_total = [897, 1749, 2624, 3537, 4491, 5582, 6863, 8436, 10909, 19042]
income_disp = [797, 1544, 2255, 3008, 3810, 4679, 5753, 6992, 8884, 14563]
income_no_cons = [101, 205, 370, 529, 681, 903, 1109, 1444, 2025, 4478]
age_head = [68.2, 63.1, 57.3, 54.6, 53.2, 51.5, 51.4, 50.1, 50.9, 51.5]
asset_total = [12407, 21970, 27511, 32508, 41889, 45692, 57223, 66604, 86386, 155475]
asset_real = [9484, 17562, 21487, 24528, 33018, 34527, 44277, 52475, 67519, 121542]
asset_financial = [2922, 4409, 6024, 7980, 8871, 11165, 12947, 14129, 18867, 33933]
debt = [1068, 2364, 3975, 5042, 6969, 8346, 9954, 12688, 15637, 25644 ]

income_total = [round(i/100,1) for i in income_total]
asset_total = [round(i/100,1) for i in asset_total]
debt = [round(i/100,1) for i in debt]
# Deciles
deciles = np.arange(1, 11)
# Plot 1: Total Income
plt.figure(figsize=(10, 5))
plt.bar(deciles, income_total, color='yellow')
plt.xlabel('Income Decile')
plt.ylabel('million won')
plt.title('Total Income(2022)')
plt.xticks(deciles)
for decile, income in zip(deciles, income_total):
    plt.text(decile, income, str(income), ha='center', va='bottom', color='black', fontsize=8)
plt.show()


# Plot 2: Share of Disposable Income and Non-Consumption Income
# Calculate the proportion of Disposable Income and Non-Consumption Income for each decile

#   
total_income = np.array(income_disp) + np.array(income_no_cons)

#  
prop_disp = np.array(income_disp) / total_income
prop_no_cons = np.array(income_no_cons) / total_income

#   
plt.figure(figsize=(10, 8))
width = 0.4  #   

plt.bar(deciles - width/2, prop_disp, width, label='Disposable Income', color='lightcoral')
plt.bar(deciles + width/2, prop_no_cons, width, label='Non-Consumption Income', color='lightgreen')
for decile, v in zip(deciles, prop_disp):
    plt.text(decile, v, round(round(float(v),3)*100,1), ha='right', va='bottom', color='black', fontsize=8)
for decile, v in zip(deciles, prop_no_cons):
    plt.text(decile, v, round(round(float(v),3)*100,1), ha='left', va='bottom', color='black', fontsize=8)


plt.xlabel('Income Decile')
plt.ylabel('%')
plt.title('Shares out of Total Income')
#X  
plt.xticks(deciles)
# Y  
plt.yticks(np.arange(0, 1.1, 0.1), [i for i in range(0,110,10)])
plt.legend()
plt.show()
# Plot 3: Total Asset Holding

plt.bar(deciles, asset_total, color='purple', label='Total Assets')
plt.xlabel('Income Decile')
plt.ylabel('million won')
plt.title('Total Asset Holdings(2022)')
plt.xticks(deciles)

for decile, asset in zip(deciles, asset_total):
    plt.text(decile, asset, str(asset), ha='center', va='bottom', color='black', fontsize=8)

plt.show()

# Plot 4: Share of Financial and Real Asset out of Total Assets
# Calculate the proportions of financial and real assets out of total assets
#   
total_asset = np.array(asset_financial) + np.array(asset_real)

#  
prop_f = np.array(asset_financial) / total_asset
prop_r = np.array(asset_real) / total_asset

#   
plt.figure(figsize=(10, 8))
width = 0.4  #   

plt.bar(deciles - width/2, prop_f, width, label='Financial Asset', color='blue')
plt.bar(deciles + width/2, prop_r, width, label='Real Asset', color='yellow')
for decile, v in zip(deciles, prop_f):
    plt.text(decile, v, round(round(float(v),3)*100,1), ha='right', va='bottom', color='black', fontsize=8)
for decile, v in zip(deciles, prop_r):
    plt.text(decile, v, round(round(float(v),3)*100,1), ha='left', va='bottom', color='black', fontsize=8)


plt.xlabel('Income Decile')
plt.ylabel('%')
plt.title('Shares out of Total Asset')
#X  
plt.xticks(deciles)
# Y  
plt.yticks(np.arange(0, 1.1, 0.1), [i for i in range(0,110,10)])
plt.legend()
plt.show()

# Plot 5: Age of the Head of Household by Income Decile
plt.figure(figsize=(10, 5))
plt.bar(deciles, age_head, color='purple')
plt.xlabel('Income Decile')
plt.ylabel('Age of Head of Household (years)')
plt.title('Age of the Head of Household by Income Decile (2022)')
plt.xticks(deciles)

# Annotate data points
for i, age in enumerate(age_head):
    plt.annotate(f'{age:.1f}', (deciles[i], age), textcoords="offset points", xytext=(0, 10), ha='center')

plt.show()

# Plot 6: Total Debt
plt.figure(figsize=(10, 5))
plt.bar(deciles, debt, color='lightcoral')
plt.xlabel('Income Decile')
plt.ylabel('million won')
plt.title('Total Debt by Income Decile (2022)')
plt.xticks(deciles)
for decile, d in zip(deciles, debt):
    plt.text(decile, d, str(d), ha='center', va='bottom', color='black', fontsize=8)
plt.show()
plt.show()

```
